## 0️⃣ What exactly is LlamaIndex? 🤖📚

**LlamaIndex** is a Python framework that helps you build applications powered by large language models (LLMs).

It provides a toolkit to ingest and organize data, retrieve the right context, interact with LLMs, augment them with your own knowledge, and create or orchestrate agentic applications. ⚙️✨
